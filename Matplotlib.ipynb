```python
# satcked bar chart
plt.bar(df['batsman'],df['2015'])
plt.bar(df['batsman'],df['2016'],bottom=df['2015'])
plt.bar(df['batsman'],df['2017'])
```

```python
# Step 2: Set width of each bar
bar_width = 0.25

# Step 3: Set positions of bars on the X-axis
r1 = np.arange(len(df['batsman']))
r2 = [x + bar_width for x in r1]
r3 = [x + bar_width for x in r2]

# Step 4: Plot the bars
plt.figure(figsize=(10, 6))
plt.bar(r1, df['2015'], color='b', width=bar_width, label='2015')
plt.bar(r2, df['2016'], color='g', width=bar_width, label='2016')
plt.bar(r3, df['2017'], color='r', width=bar_width, label='2017')

# Step 5: Customize the chart
plt.xlabel('Batsman', fontweight='bold')
plt.ylabel('Runs', fontweight='bold')
plt.title('Runs Scored by Batsmen (2015-2017)')
plt.xticks([r + bar_width for r in range(len(df['batsman']))], df['batsman'], rotation=45)
plt.legend()

plt.tight_layout()
plt.show()
```

# 3D Surface Plots

```python
x = np.linspace(-10,10,100)
y = np.linspace(-10,10,100)
```

```python
xx, yy = np.meshgrid(x,y)
(100, 100)
```

```python
z = xx**2 + yy**2
z.shape

(100, 100)
```

```python
fig = plt.figure(figsize=(12,8))

ax = plt.subplot(projection='3d')

p = ax.plot_surface(xx,yy,z,cmap='viridis')
fig.colorbar(p)
```

```python
z = np.sin(xx) + np.cos(yy)
fig = plt.figure(figsize=(12,8))

ax = plt.subplot(projection='3d')

p = ax.plot_surface(xx,yy,z,cmap='viridis')
fig.colorbar(p)
```

# Contour Plots
